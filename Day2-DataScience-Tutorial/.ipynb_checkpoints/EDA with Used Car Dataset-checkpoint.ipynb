```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pylab as plot
import seaborn as sns


df = pd.read_csv('vehicles.csv', low_memory = False)
## Due to memory error, I have deleted about 97% of the data. 
df.head() # Returns the first 5 rows + headers
df.describe()
```

```python
## As seen above, the values are in scientific format with the e+09. 
## We want to convert the scientific format to general formula
## Hence we use this function. Read more 
## https://stackoverflow.com/questions/40347689/dataframe-describe-suppress-scientific-notation/47207283

df.describe().apply(lambda s: s.apply(lambda x:format(x, 'f')))

```

## Cleaning Dataset


```python
# Remove redundant variables
df_cleaned = df.copy()

df_cleaned = df_cleaned.copy().drop(['url', 'image_url', 'region_url'], axis=1)
```

```python
# Selecting variables
# Removing columns with too many null values
Na_val = df_cleaned.isna().sum() 

## Only select the usable columns who have less that 30% of its data as null values

def na_filter(na, threshold = 0.3):
    col_pass = []
    for i in na.keys():
        if na[i]/df_cleaned.shape[0] < threshold:
            col_pass.append(i)
    return col_passdf_cleaned == df_cleaned[na_filter(NA_val)]
df_cleaned.columns
```

```python
# Removing outliers
df_cleaned = df_cleaned[df_cleaned['price'].between(999.99, 99999.00)]
df_cleaned = df_cleaned[df_cleaned['year'] > 1990]
df_cleaned = df_cleaned[df_cleaned['odometer'] < 85000.00]
df_cleaned.describe().apply(lambda s:s.apply(lambda x:format(x, 'f')))
```

## Correlation Matrix - Heatmap
A correlation matrix is a table that shows the correlation coefficients between many variables. I used sns.heatmap() to plot a correlation matrix of all of the variables in the used car dataset.

Read more about pandas corr function [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html)
Read more about seaborn heatmap plot [here](https://seaborn.pydata.org/generated/seaborn.heatmap.html)

```python
# Calculate correlation matrix
corr = df_cleaned.corr() 
## Plot map
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

```python
df_cleaned
```

```python
# More datacleaning. Drop id and county column
df_cleaned = df_cleaned.drop(['id', 'county'], axis=1, inplace=True)
df_cleaned
```

```python
# Replot Heatmap
# Calculate correlation matrix
corr = df_cleaned.corr() 
## Plot map
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

## Histogram 

Heatmaps and Scatterplots are for analyzing between two variables
Histogram is for 1 variable. It shows the distribution of that variable's set of values


```python

df_cleaned['odometer'].plot(kind = 'hist', bins=50, figsize=(12,6), facecolor='grey', edgecolor='black', title='Odometer Values')
```

```python

df_cleaned['year'].plot(kind = 'hist', bins=50, figsize=(12,6), facecolor='grey', edgecolor='black', title='Year Values')
```
